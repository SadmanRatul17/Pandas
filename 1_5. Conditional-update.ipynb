```python
import pandas as pd
```

```python
df = pd.read_excel('ResultGrade.xlsx')
df
```

```python
df['Total'] = df['Math']+df['English']+df['Bangla']+df['Science']+df['History']
df
```

```python
df['percent'] = df['Total'] / 500 * 100
df
```

```python
df["Grade"]=None
df
```

```python
# we want under 50 pecent rows

df[df['percent'] < 50]
```

```python
# another way to find
df.loc[df['percent'] < 50]
```

```python
# add FAIL in grade that has percent under 50 

df.loc[df['percent'] < 50,['Grade']]='FAIL'
df
```

```python
df.loc[df['percent'] > 80,['Grade']]='A+'
df
```
