## Conditional delete from a Dataframe 

```python
df = pd.read_excel('ResultGrade.xlsx')
df
```

```python
# we want to delete 7 index row. (Temporary delete)
# if you want to delete permanently then use df.drop(index=7,inplace=True) 

df.drop(index=7) 
```

```python
# delete a column . (Temporary delete)

df.drop(columns=df.columns[4]) 
```

```python
df
```

```python
df['Total'] = df['Math']+df['English']+df['Bangla']+df['Science']+df['History']
df
```

```python
df['percent'] = df['Total'] / 500 * 100
df
```

```python
df[df['percent'] > 80]
```

```python
df[df['percent'] > 80].index
```

```python
df.drop(index=df[df['percent'] > 80].index,inplace=True) #permanent delete
df
```
